```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $\Delta L(k)$', fontsize=18)
#plt.xlim(xmin=0, xmax=0.03)
#plt.ylim(ymin=0, ymax=100)
plt.ylabel('$\Delta L$, мм', fontsize=20, rotation=0, ha='right')
plt.xlabel('$k$', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)



yerr_1 = 1
xerr_1 = 2



#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    k_3kgz,
    L_3kgz,
    fmt='ko',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='3 kHz',
    #xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)

plt.errorbar(
    k_4kgz,
    L_4kgz,
    fmt='ro',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='4 kHz',
    #xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)

plt.errorbar(
    k_5kgz,
    L_5kgz,
    fmt='bo',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='5 kHz',
    #xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)

plt.errorbar(
    k_6kgz,
    L_6kgz,
    fmt='yo',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='6 kHz',
    #xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)

z_1 = np.polyfit(k_3kgz, L_3kgz, deg=1)
polynom_1 = np.poly1d(z_1)
plt.plot(k_6kgz, polynom_1(k_6kgz),"k--", linewidth=1)

z_2 = np.polyfit(k_4kgz, L_4kgz, deg=1)
polynom_2 = np.poly1d(z_2)
plt.plot(k_6kgz, polynom_2(k_6kgz),"r--", linewidth=1)

z_3 = np.polyfit(k_5kgz, L_5kgz, deg=1)
polynom_3 = np.poly1d(z_3)
plt.plot(k_6kgz, polynom_3(k_6kgz),"b--", linewidth=1)

z_4 = np.polyfit(k_6kgz, L_6kgz, deg=1)
polynom_4 = np.poly1d(z_4)
plt.plot(k_6kgz, polynom_4(k_6kgz),"y--", linewidth=1)


plt.legend(fontsize=18, markerscale = 1)
plt.show()

plot_1.savefig(
    'L(k).pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
