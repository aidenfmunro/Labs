```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
p_23 = np.array([
3.9,
3.7,
3.2 ,
2.8 ,
2.3])

u_23 = np.array([
137,
128,
107,
88,
59])

p_33 = np.array([
4.1,
3.6 ,
3.4,
3.1,
2.5])

u_33 = np.array([
135,
114,
105,
89 ,
64])

p_43 =np.array([
4.1,
3.0,
2.0])

u_43 = np.array([
110,
68,
38])

```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $\Delta T(\Delta p)$', fontsize=18)
#plt.xlim(xmin=0, xmax=0.03)
#plt.ylim(ymin=0, ymax=100)
plt.ylabel('$\Delta T$, $^\circ C$', fontsize=20, rotation=0, ha='right')
plt.xlabel('$\Delta p$, бар', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)



yerr_1 = 3
xerr_1 = 0.1



#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    p_23,
    u_23/40.7,
    fmt='ko',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='23 $^\circ C$',
    xerr=xerr_1,
    yerr=yerr_1/38.9,
    capsize=3
)

plt.errorbar(
    p_33,
    u_33 / 41.6,
    fmt='ro',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='33 $^\circ C$',
    xerr=xerr_1,
    yerr=yerr_1 / 41.6,
    capsize=3
)

plt.errorbar(
    p_43,
    u_43 / 42.5,
    fmt='bo',
    linewidth=0,
    markersize=8,
    elinewidth=1,
    label='43 $^\circ C$',
    xerr=xerr_1,
    yerr=yerr_1 / 42.5,
    capsize=3
)


z_1 = np.polyfit(p_23, u_23/40.7, deg=1)
polynom_1 = np.poly1d(z_1)
plt.plot(p_23, polynom_1(p_23),"k--", linewidth=1)

z_2 = np.polyfit(p_33, u_33 / 41.6, deg=1)
polynom_2 = np.poly1d(z_2)
plt.plot(p_33, polynom_2(p_33),"r--", linewidth=1)

z_3 = np.polyfit(p_43, u_43 / 42.5, deg=1)
polynom_3 = np.poly1d(z_3)
plt.plot(p_43, polynom_3(p_43),"b--", linewidth=1)


plt.legend(fontsize=18, markerscale = 1)
plt.show()

plot_1.savefig(
    'DPDT.pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
