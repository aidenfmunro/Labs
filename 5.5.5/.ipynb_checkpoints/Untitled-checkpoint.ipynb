```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_spectrum(table, titlename):
    spectrum = pd.read_csv(table, skiprows=5)
    filename, _ = os.path.splitext(table)

    fig, ax = plt.subplots()

    plt.plot(spectrum['Channel'], spectrum['Counts'])
    ax.grid()
    
    ax.set_title(titlename)
    ax.set_xlabel(r'Номер канала $N$')
    ax.set_ylabel(r'Количество зафиксированных частиц')

    plt.savefig('img/' + filename + '.png')
    plt.show()
```

```python
plot_spectrum('Bgrd.csv', 'Фон')
```

```python
plot_spectrum('Co.csv', r'$^{60}$Co')
```

```python
plot_spectrum('Na.csv', r'$^{22}$Na')
```

```python
plot_spectrum('Cs.csv', r'$^{137}$Cs')
```

```python
plot_spectrum('Am.csv', r'$^{241}$Am')
```

```python
plot_spectrum('Eu.csv', r'$^{152}$Eu')
```
